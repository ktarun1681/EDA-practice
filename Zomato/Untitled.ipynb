We can combine the two dataset based on the common column in the two datasets. As we can see there is country code common in both the datasets. 

```python
final_df =pd.merge(df, df_country, on='Country Code', how='left' )
```

```python
final_df.head()
```

From the above, code we can see that how many datapoints are there for a particular country. As we can see there are 8652 values are there for India which is the maximum. 

```python
country_name = final_df.Country.value_counts().index
```

```python
country_name
```

```python
country_values = final_df.Country.value_counts().values
```

## How to draw the pie chart. 

```python
plt.pie(country_values[:3], labels=country_name[:3])

#showing the top 3 countries in the pie chart which uses zomato. 
```

If We don't use the top 3 then we are not able to see the data properly as with respect to India the values for the other countries is very less. 

```python
plt.pie(country_values[:3], labels=country_name[:3], autopct='%1.2f%%') #using pct we can also see the percentages of the values too. 
```

Around 94 percent of the data values(transactions)  are from india, and after that US and UK.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation:

1. 4.5 to 4.9 ---> Excellent Rating
2. 4.0 to 4.4 ---> Very Good Rating
3. 2.5 to 3.4 ---> Average Rating
4. 1.0 to 2.4 ---> Poor Rating
5. 0          ---> Not Rated

Majortiy of the people have not rated. 

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15,6)
sns.barplot(x='Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
matplotlib.rcParams['figure.figsize'] = (15,6)
sns.barplot(x='Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette = ['blue', 'red', 'orange', 'yellow' 'green', 'darkgreen'])
```

### Observations:

1. Not Rated count is very high. 
2. Maximum number of rating are in between 2.5 to 3.4

As the maximum number of rating is in between 2.5 to 3.4 . So, we can take the ones who have not rated the values to be in this range. 

```python
#count plot

sns.countplot(x = 'Rating color', data = ratings, palette = ['blue', 'red', 'orange', 'yellow' 'green', 'green'])
```

This countplot indicates that the orange color is most frequent that means the ratings are highest in between this range.
